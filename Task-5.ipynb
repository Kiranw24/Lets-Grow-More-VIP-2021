```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

* Read Dataset Using Pandas library

```python
df_matches = pd.read_csv(r'K:\\machine learning\\sparks foundation internship\\matches.csv')
df_matches
```

* Here we Check Shape of our givendatset

```python
df_matches.shape # (756 rowss & 18 columns)
```

* Check total null values in each column.

```python
df_matches.isnull().sum()
```

* umpire3 column has most number of null values

```python
df_matches['winner'] = df_matches['winner'].fillna(method='ffill')
```

# Explorating Data Analysis

* Here we plot Total number of matches played in each IPL season(2008-19) 

```python
bg_color = (0.25, 0.25, 0.25)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"orange",
            "xtick.color":"white",
            "ytick.color":"white",
            "axes.labelcolor":"orange",
            "axes.grid":False,
            'axes.labelsize':80,
            'figure.figsize':(25,16),
            'xtick.labelsize':15,
            'ytick.labelsize':15})

plt.figure(figsize=(15,7))
sns.countplot(x='season',
              data=df_matches,
              order = df_matches['season'].value_counts().index.sort_values()
              )
plt.xlabel('Season(2008-19)',fontsize=18)
plt.ylabel('Count',fontsize=18)
plt.title('Number of matches played in each IPL season',fontsize=22)
plt.show()
```

* Here we conclude that 

* The highest number of matches was played in the 2013 season, followed by the highest number of matches in the 2012 and 2011 seasons

```python
# Here one mistacke is there during making datset in 'winner' column.
df_matches['winner'] = df_matches['winner'].replace('Rising Pune Supergiant','Rising Pune Supergiants')
```

* We are plotting graph here to see which team won how many matches.

```python
bg_color = (0.25, 0.25, 0.25)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"orange",
            "xtick.color":"white",
            "ytick.color":"white",
            "axes.labelcolor":"orange",
            "axes.grid":False,
            'axes.labelsize':80,
            'figure.figsize':(25,16),
            'xtick.labelsize':18,
            'ytick.labelsize':18})

plt.figure(figsize=(15,7))
sns.countplot(y='winner',
              data=df_matches,
              order = df_matches['winner'].value_counts().index,
              )
plt.xlabel('No. of Matches Win',fontsize=20)
plt.ylabel('Team',fontsize=20)
plt.title('Number of matches win by each team',fontsize=22)
```

* we conclude that 'Mumbai Indians' win most number of matches, followed by 'chennai super kings'.

* We are plotting graph here to see how many matches were played on which ground.

```python
bg_color = (0.25, 0.25, 0.25)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"orange",
            "xtick.color":"white",
            "ytick.color":"white",
            "axes.labelcolor":"orange",
            "axes.grid":False,
            'axes.labelsize':80,
            'figure.figsize':(25,20),
            'xtick.labelsize':18,
            'ytick.labelsize':18})
plt.figure(figsize=(15,7))
sns.countplot(y='venue',
              data=df_matches,
              order = df_matches['venue'].value_counts().iloc[:10].index,
              )
plt.xlabel('No. of Matches played on That stadium',fontsize=20)
plt.ylabel('Name of The Stadium',fontsize=20)
plt.title('Number of Played on each Stadium',fontsize=22)
```

* Most matches were played at 'Eden Gardens' Stadium, followed by 'M Chinnaswamy Stadium'.

* Here we plot pie-plot to see if Toss win then how many times captain choose field first or Bat first in percentages.

```python
bg_color = (0.25, 0.25, 0.25)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"white",
            "xtick.color":"orange",
            "ytick.color":"orange",
            "axes.labelcolor":"white",
            "axes.grid":False,
            'axes.labelsize':80,
            'figure.figsize':(25,20),
            'xtick.labelsize':25,
            'ytick.labelsize':20})

plt.figure(figsize=(6,6))
sizes = df_matches.toss_decision.value_counts()
labels = df_matches.toss_decision.value_counts().index
plt.pie(sizes,colors = ['black','blue'],
         labels=labels,
         autopct='%1.1f%%',
         startangle=90,
         pctdistance=0.75,
         explode = (0.025,0.025),textprops={'size': 'x-large'})

#draw white circle
centre_circle = plt.Circle((0,0),0.60,fc='gray')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('If Toss Win Then Team choose',fontsize=15)
plt.show()
```

* After winning the toss, 61.2% of the time the captain chose fielding first.

* Plot the Graph which show Team win the toss and also win the Match.

```python
df_matches['result'].value_counts()
```

```python
bg_color = (0.25, 0.25, 0.25)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"white",
            "xtick.color":"orange",
            "ytick.color":"orange",
            "axes.labelcolor":"white",
            "axes.grid":False,
            'axes.labelsize':80,
            'figure.figsize':(25,20),
            'xtick.labelsize':25,
            'ytick.labelsize':20})

plt.figure(figsize=(25,10))
sns.countplot(x='result',
              hue='toss_winner',
              data=df_matches,
              order = df_matches['result'].value_counts().index.sort_values(),palette='Accent'
              )
plt.xlabel('Result',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.legend()
plt.title('one Team win the toss and also win the match ', fontsize=16)
plt.show()
```

* Mumbai Indians win the toss and also win the match most of the times

* Here we plot a graph to see how many times the captain choosed bat or field first in each season.

```python
plt.figure(figsize=(25,10))
sns.countplot(x='season',
              hue='toss_decision',
              data=df_matches,
              order = df_matches['season'].value_counts().index.sort_values(),
              palette='twilight_shifted')
plt.xlabel('Season(2008-19)',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.legend(fontsize='xx-large')
plt.title('Decision to field or bat in each IPL season', fontsize=16)
plt.show()
```

* One thing we can conclude from the graph is that there was not much difference before the year of 2016 but there is a big difference during the selections in 2016 and beyond.
* In 2016, 17, 18, 19, the decision to field first was taken many times.

* plot the barplot to see Top 10 players who have won the Man of the Match award most of the time.

```python
bg_color = (0.25, 0.25, 0.25)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"white",
            "xtick.color":"orange",
            "ytick.color":"orange",
            "axes.labelcolor":"white",
            "axes.grid":False,
            'axes.labelsize':80,
            'figure.figsize':(25,20),
            'xtick.labelsize':12,
            'ytick.labelsize':15})

plt.figure(figsize=(15,7))
player_of_match=df_matches['player_of_match'].value_counts()[:10]
sns.barplot(player_of_match.index,
            player_of_match.values,
            palette='Accent')

plt.title("Top 10 players who have won the Man of the Match award most of the time (2008-19)",fontsize=22)
plt.xlabel('Player',fontsize=17)
plt.ylabel('Count',fontsize=17)
plt.show()
```

* Chris Gayle is in first place and ABD is in second place.
