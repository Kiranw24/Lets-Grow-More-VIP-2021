# Exploratory Data Analysis

* Here we plot pairplot using seaborn library for visualize the data

```python
sns.pairplot(df)
```

```python
df.corr() #here we check correlation between the 'sales','quantity','discount' and 'profit'
```

* here we plot heatmap to visualize this correlation in better way using seaborn library which we import previously

```python
sns.heatmap(df.corr(),cmap='rocket_r',annot=True)
```

* here in heatmap we see the dark portion show more correlation and light portion show less correlation
* from above we conclude that
    * profit is more-moderate correlate with Sales
    * also profit is moderate correlate with quantity
    * and profit is negatively correlate with the discount

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'],df['Profit'],data=df)
```

* here we get relation between Discount and Profit
    * Here we see around less than 25% discount gives a profit but after 25% we are in loss.

# Category wise relation with profit and sales

```python
plt.figure(figsize=(7,7))
plt.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,textprops={'fontsize':15},autopct='%0.1f%%',
       shadow=True,startangle = 90)
plt.title('Category',fontsize=20)
plt.show()
```

* from above we see 60.3% contribution of 'Office Supplies'.
*Furniture on the second position

```python
df.groupby(['Category'])[['Profit','Sales']].sum().plot.bar()
plt.ylabel('profit or no. of sales')
plt.legend(loc=2)
plt.show()
```

* Technology gives more profit and office supplies also give much profit
* we see furniture having more sales but getting profit is very less

# Sub-category wise relation with the profit and sales

```python
df.groupby(['Sub-Category'])[['Profit','Sales']].sum().plot.bar()
plt.ylabel('profit or no. of sales')
plt.legend(loc=2)
plt.show()
```

Here we not see that much confortable now we have another method to plot

```python
from matplotlib import style
style.use('ggplot')
subcategory_sales = df.groupby('Sub-Category').Sales.sum()
subcategory_profit = df.groupby('Sub-Category').Profit.sum()

subcategory_sales.plot(kind='bar',figsize=(10,5),fontsize=15,color='g')
subcategory_profit.plot(kind='bar',figsize=(10,5),fontsize=15,color='y')

plt.xlabel('Sub-Category',fontsize=20)
plt.ylabel('profit or no. of sales',fontsize=20)
plt.legend()
plt.show()
```

* from the above barplot we can see that tables giving us loss.
    * we see Copies, accessories gives maximum profit
    * also phones gives more profit

# Region-wise relation with profit and sales

```python
plt.figure(figsize=(7,7))
plt.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,textprops={'fontsize':15},autopct='%0.1f%%',
       shadow=True,startangle = 90)
plt.title('Region',fontsize=20)
plt.show()
```

* from above we see the west and east region has more important to us.

```python
df.groupby(['Region'])[['Profit','Sales']].sum().plot.bar()
plt.ylabel('profit or no. of sales')
plt.legend(loc=2)
plt.show()
```

* West region gives more profit
but we cant restrict any of the region beacuse every region has something amount of contribution in profit

# State-Wise Relation with Profit and sales

```python
df.groupby(['State'])[['Profit','Sales']].sum().plot.bar(figsize=(16,9))
plt.ylabel('profit or no. of sales')
plt.legend(loc=1)
plt.show()
```

* Texas, Ohio and Illinois are the States where overall sales are in loss.
